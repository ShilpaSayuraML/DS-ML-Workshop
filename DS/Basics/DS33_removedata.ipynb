#How to Mark and Remove Missing Data

```python
diabetes_url="https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
# https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.names

# load and summarize the dataset
from pandas import read_csv
# load the dataset
df = read_csv(diabetes_url, header=None)
# summarize the dataset
print(df.describe())
```

some columns have a minimum value of zero (0).
They indicate presence of an invalid or missing values.
The following columns have invalid zero minimum value:

* 1: Plasma glucose concentration
* 2: Diastolic blood pressure
* 3: Triceps skinfold thickness
* 4: 2-Hour serum insulin
* 5: Body mass index

```python
print(df.head(20))
#we can clearly see 0 values in the columns 2, 3, 4, and 5
```

```python
count = (df[2] == 0).sum()
print('Count of zeros in Column  2 : ', count)
```

```python
from pandas.core.internals.construction import dataclasses_to_dicts
#filter out rows that contain zero values
df2 = df.loc[(df != 0).all(axis=1), :].head(10)
df2
```

count of the number of missing values on each of the columns

```python
# count the number of missing values for each column
num_missing = (df[[1,2,3,4,5]] == 0).sum()
# report the results
print(num_missing)
```

replace 0s with Nan

```python
from numpy import nan
df[[1,2,3,4,5]] = df[[1,2,3,4,5]].replace(0, nan)
# count the number of nan values in each column
print(df.isnull().sum())
```

```python
df.head(20)
```

Having missing values in a dataset can cause errors with some ML algorithms.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
# split dataset into inputs and outputs
values = df.values
X = values[:,0:8]
y = values[:,8]
# define the model
model = LinearDiscriminantAnalysis()
# define the model evaluation procedure
cv = KFold(n_splits=3, shuffle=True, random_state=1)
# evaluate the model
result = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
# report the mean performance
print('Accuracy: %.3f' % result.mean())
```

Remove Rows With Missing Values

```python
# drop rows with missing values
df.dropna(inplace=True)
# split dataset into inputs and outputs
values = df.values
X = values[:,0:8]
y = values[:,8]
# define the model
model = LinearDiscriminantAnalysis()
# define the model evaluation procedure
cv = KFold(n_splits=3, shuffle=True, random_state=1)
# evaluate the model
result = cross_val_score(model, X, y, cv=cv, scoring='accuracy')
# report the mean performance
print('Accuracy: %.3f' % result.mean())
```
