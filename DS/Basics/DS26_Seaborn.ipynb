#Seaborn 
Seaborn is a library that uses Matplotlib underneath to plot graphs. 
It will be used to visualize random distributions.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show()
```

Binomial Distribution


```python
#Given 10 trials for coin toss generate 10 data points:

from numpy import random

x = random.binomial(n=1, p=0.5, size=10)
x
```

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

normal distribution is continous whereas binomial is discrete, but if there are enough data points it will be quite similar to normal distribution with certain loc and scale.

```python
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns

sns.displot(random.normal(loc=50, scale=5, size=1000), kind="kde")
sns.displot(random.binomial(n=100, p=0.5, size=1000,), kind="kde", color="red")

plt.show()
```

```python
#empirical cumulative distribution functions (ECDFs):

sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```

```python
#histogram mode, it is also possible to add a KDE curve:

sns.displot(data=penguins, x="flipper_length_mm", kde=True)
```

```python
#bivariate plot, assign both x and y:

sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
```

```python
#subplots
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="sex", kind="kde")
```

```python
sns.displot(
    data=penguins, y="flipper_length_mm", hue="sex", col="species",
    kind="ecdf", height=4, aspect=.7,
)
```

```python
g = sns.displot(
    data=penguins, y="flipper_length_mm", hue="sex", col="species",
    kind="kde", height=4, aspect=.7,
)
g.set_axis_labels("Density (a.u.)", "Flipper length (mm)")
g.set_titles("{col_name} penguins")
```

```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)

sinplot()
```
