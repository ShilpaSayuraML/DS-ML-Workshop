```python
from sklearn.linear_model import LinearRegression
```

```python
n = 10
x = np.arange(n)
y = 4*x-9 + (5*np.random.randn(n)) # y = f(x) : f(x) = 4x-9 : g(x) = 3x+6
plt.scatter(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures

m = LinearRegression()
m.fit(x[:,np.newaxis],y)
y_pred = m.predict(x[:,np.newaxis])

pf = PolynomialFeatures(degree=10)
x_poly = pf.fit_transform(x[:,np.newaxis])
m2 = LinearRegression()
m2.fit(x_poly,y)
y_pred_poly = m2.predict(x_poly)
plt.scatter(x,y)
#plt.plot(x,y_pred,color = 'red')
plt.plot(x,y_pred,color = 'red')
plt.plot(x,y_pred_poly,color = 'green')
```
