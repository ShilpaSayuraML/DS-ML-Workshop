```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Generating example data
np.random.seed(42)
hours_studied = np.random.uniform(1, 10, 100)
pass_fail = np.where(hours_studied > 5, 1, 0)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(hours_studied.reshape(-1, 1), pass_fail, test_size=0.2, random_state=42)

# Creating and training the logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Making predictions on the test set
predictions = model.predict(X_test)

# Plotting the data and decision boundary
plt.scatter(hours_studied, pass_fail, color='blue', label='Actual data')
plt.plot(np.sort(X_test, axis=0), model.predict_proba(np.sort(X_test, axis=0).reshape(-1, 1))[:, 1], color='red', label='Decision boundary')
plt.xlabel('Hours Studied')
plt.ylabel('Pass (1) / Fail (0)')
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_diabetes
from sklearn.metrics import accuracy_score, confusion_matrix

# Load diabetes dataset
diabetes = load_diabetes()
X_diabetes = diabetes.data[:, np.newaxis, 2]  # Using only one feature for simplicity
y_diabetes = (diabetes.target > diabetes.target.mean()).astype(int)

# Splitting the data into training and testing sets
X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(X_diabetes, y_diabetes, test_size=0.2, random_state=42)

# Creating and training the logistic regression model
model_diabetes = LogisticRegression()
model_diabetes.fit(X_train_diabetes, y_train_diabetes)

# Making predictions on the test set
predictions_diabetes = model_diabetes.predict(X_test_diabetes)

# Evaluating the model
accuracy_diabetes = accuracy_score(y_test_diabetes, predictions_diabetes)
conf_matrix_diabetes = confusion_matrix(y_test_diabetes, predictions_diabetes)

print(f"Diabetes Dataset - Accuracy: {accuracy_diabetes:.2f}")
print("Confusion Matrix:")
print(conf_matrix_diabetes)
```

```python
import matplotlib.pyplot as plt

# Visualizing the logistic regression decision boundary
plt.scatter(X_diabetes, y_diabetes, color='blue', label='Actual data')
plt.plot(X_test_diabetes, model_diabetes.predict(X_test_diabetes.reshape(-1, 1)), color='red', linewidth=3, label='Decision boundary')
plt.xlabel('Feature (Diabetes Dataset)')
plt.ylabel('Target (1 or 0)')
plt.title('Logistic Regression Decision Boundary - Diabetes Dataset')
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer

# Load breast cancer dataset
cancer = load_breast_cancer()
X_cancer = cancer.data
y_cancer = cancer.target

# Splitting the data into training and testing sets
X_train_cancer, X_test_cancer, y_train_cancer, y_test_cancer = train_test_split(X_cancer, y_cancer, test_size=0.2, random_state=42)

# Creating and training the logistic regression model with increased iterations
model_cancer = LogisticRegression(max_iter=1000)  # Set max_iter to your desired number
model_cancer.fit(X_train_cancer, y_train_cancer)

# Making predictions on the test set
predictions_cancer = model_cancer.predict(X_test_cancer)

# Evaluating the model
accuracy_cancer = accuracy_score(y_test_cancer, predictions_cancer)
conf_matrix_cancer = confusion_matrix(y_test_cancer, predictions_cancer)

print(f"Breast Cancer Dataset - Accuracy: {accuracy_cancer:.2f}")
print("Confusion Matrix:")
print(conf_matrix_cancer)
```
