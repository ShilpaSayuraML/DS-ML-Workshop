```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random x values
np.random.seed(42)
x = np.random.rand(50) * 10  # 50 random values between 0 and 10

# Generate corresponding y values with some noise
m = 2  # slope
b = 1  # y-intercept
noise = np.random.normal(0, 1, 50)  # adding some random noise
y = m * x + b + noise

# Perform linear regression to find the best-fit line
fit_params = np.polyfit(x, y, 1)
fit_line = np.poly1d(fit_params)

# Create a scatter plot
plt.scatter(x, y, label='Linear Regression Data')

# Plot the regression line
plt.plot(x, fit_line(x), color='red', label='Regression Line: y = {:.2f}x + {:.2f}'.format(*fit_params))

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

```python
# Fitted parameters from linear regression
m = fit_params[0]  # slope
b = fit_params[1]  # y-intercept

# Inference for x = 4.5
inferred_y = m * 4.5 + b

print("Inferred y for x = 4.5:", inferred_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 100)

# Generate y values with a quadratic term for the "U" shape
y = 2 * x**2 + 3 * x + 1 + np.random.normal(0, 5, size=len(x))  # Adding some random noise

# Perform linear regression
coefficients = np.polyfit(x, y, 2)
regression_line = np.polyval(coefficients, x)

# Create a scatter plot
plt.scatter(x, y, label='Data Points')

# Plot the regression line
plt.plot(x, regression_line, color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

```python
# Use the coefficients to calculate the y value for x = 4.5
inferred_y = np.polyval(coefficients, 2.5)

print("For x = 2.5, the inferred y is:", inferred_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 2 * np.pi, 100)
# Generate y values with a sinusoidal shape and add some random noise
y = np.sin(x) + np.random.normal(0, 0.3, size=len(x))

# Perform linear regression on transformed features
X_transformed = np.column_stack((x, np.sin(x)))
coefficients = np.linalg.lstsq(X_transformed, y, rcond=None)[0]

# Create a scatter plot
plt.scatter(x, y, label='Data Points')

# Plot the linear regression line
regression_line = coefficients[0] * x + coefficients[1] * np.sin(x)
plt.plot(x, regression_line, color='red', label='Linear Regression')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.show()
```

```python
# Use the coefficients to calculate the y value for x = 2.7
inferred_y = coefficients[0] * 2.7 + coefficients[1] * np.sin(2.7)

print("For x = 2.7, the inferred y is:", inferred_y)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Generate random data with two independent variables
np.random.seed(42)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
noise = np.random.randn(100, 1)
y = 4 + 2 * X1 + 3 * X2 + noise

# Combine independent variables into a feature matrix
X = np.column_stack((X1, X2))

# Fit a multivariate linear regression model
model = LinearRegression()
model.fit(X, y)

# Coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

# Print coefficients and intercept
print("Coefficients:", coefficients)
print("Intercept:", intercept)

# Make predictions for new data
new_data = np.array([[1.5, 2.0]])  # New data with two independent variables
prediction = model.predict(new_data)

print("Prediction for new data:", prediction)

# Plotting the fitted plane
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, y, c='blue', marker='o', label='Data Points')

# Creating a meshgrid for the plane
x1_range = np.linspace(min(X1), max(X1), 20)
x2_range = np.linspace(min(X2), max(X2), 20)
x1_plane, x2_plane = np.meshgrid(x1_range, x2_range)
fitted_plane = intercept + coefficients[0][0] * x1_plane + coefficients[0][1] * x2_plane

# Plotting the fitted plane
ax.plot_surface(x1_plane, x2_plane, fitted_plane, alpha=0.5, color='red', label='Fitted Plane')
ax.set_xlabel('X1')
ax.set_ylabel('X2')
ax.set_zlabel('y')


#ax.legend(labels=['Data Points', 'Fitted Plane'])

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
california_housing = fetch_california_housing()
data = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
data['target'] = california_housing.target

# Display the first few rows of the dataset
print("Dataset Head:")
print(data.head())

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Pairplot for a quick overview of relationships between variables
sns.pairplot(data, vars=data.columns[:-1], diag_kind='kde', height=2.5)
plt.suptitle("Pairplot of California Housing Dataset", y=1.02)
plt.show()

# Correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of California Housing Dataset")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Select a feature for linear regression (e.g., 'MedInc' - median income)
feature = 'MedInc'

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[feature].values.reshape(-1, 1), data['target'], test_size=0.2, random_state=42)

# Create and fit a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
predictions = model.predict(X_test)

# Visualize the linear regression
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_test.flatten(), y=y_test, label='Actual Values')
sns.lineplot(x=X_test.flatten(), y=predictions, color='red', label='Linear Regression Line')
plt.title(f'Linear Regression on California Housing Dataset ({feature} vs. Target)')
plt.xlabel(feature)
plt.ylabel('Median House Value')
plt.legend()
plt.show()

# Calculate and print Mean Squared Error
mse = mean_squared_error(y_test, predictions)
print(f'Mean Squared Error: {mse}')
```
