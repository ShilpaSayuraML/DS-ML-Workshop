```python
import seaborn as sns
sns.get_dataset_names()
```

```python
df=sns.load_dataset("flights")
df.info()
```

```python
df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt
# Plot the number of passengers by year
plt.figure(figsize=(12, 6))
sns.lineplot(x='year', y='passengers', data=df, marker='o', color='skyblue')
plt.title('Number of Passengers Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python


# Pivot the data to create a matrix of months and years
pivot_data = df.pivot('month', 'year', 'passengers')

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=.5)
plt.title('Monthly Frequency of Passengers Over the Years')
plt.xlabel('Year')
plt.ylabel('Month')
plt.show()
```

```python

# Plot a bar plot of monthly frequency
plt.figure(figsize=(12, 6))
sns.barplot(x='month', y='passengers', data=df, errorbar=None)
plt.title('Monthly Frequency of Passengers')
plt.xlabel('Month')
plt.ylabel('Average Number of Passengers')
plt.show()
```

```python
# Plot a scatter plot of monthly frequency
plt.figure(figsize=(12, 6))
sns.scatterplot(x='year', y='passengers', data=df, color='skyblue')
plt.title('Scatter Plot of Monthly Frequency of Passengers')
plt.xlabel('Month')
plt.ylabel('Number of Passengers')
plt.show()
```
